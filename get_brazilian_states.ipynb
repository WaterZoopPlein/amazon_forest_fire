```python
import pandas as pd
import numpy as np
```

```python
link = "https://en.wikipedia.org/wiki/Federative_units_of_Brazil"

states = pd.read_html(link,header=0)[1]
```

```python
states.columns = map(str.lower, states.columns)
states.head()
```

```python
columns_rename_map = {'flag and name' : 'name', 
                      'area(km2)[1]' : 'area_km2', 
                      'population(2019)[2]' : 'population',
                      'density (perkm2, 2019)' : 'density_perkm2',
                      'gdp (r$millions, 2016)[3]' : 'gdp_r$million',
                      'hdi(2017)[4]' : 'hdi'}
states.rename(columns=columns_rename_map, inplace=True)
```

```python
states.head()
```

```python
states.to_csv('data/states.csv', index=False)
```
